## Creating Dataframes using random/dummy data

````python
## Example 1 ##
df1 = pd.DataFrame(
    np.random.rand(100, 4),
    index=range(10, 110),
    columns=list("ABCD"))

print(f'Shape: {df1.shape}')
print(f'Indices: {df1.index}')
print(f'Column names: {df1.columns}\n')
df1
````

````python
## Example 2 ##
df2 = pd.DataFrame(
    np.random.rand(9, 4),
    index=list("abcdefghi"),
    columns=list("ABCD"))

print(f'Shape: {df2.shape}')
print(f'Indices: {df2.index}')
print(f'Column names: {df2.columns}\n')
print(f'Information about dataframe:')
df2.info()
print()
df2
````

### Using lambda functions as selectors

````python
# This lambda function takes a dataframe 'df' as input and outputs those rows where the corresponding value of attribute A > 0
selector = lambda df: df['A'] > 0.5
selector
````

````python
df2.loc[selector]
````

````python
selector2 = lambda df: (df['A'] > 0.5)&(df['B'] < 0.2)
df2.loc[selector2]
````

### Using boolean conditions as selectors

- OR Operator: |
- NOT Operator: ~
- AND Operator: &

````python
condition_for_selection = (df2['A'] > 0.5) & (df2['B'] < 0.2)
print(f'Condition for selection: {list(condition_for_selection)}')
print()
print(f'Selected data: ')
df2[condition_for_selection]
````

````python
condition_for_selection2 = (df2['A'] > 0.5) | ~(df2['B'] < 0.2)
print(f'Condition for selection: {list(condition_for_selection2)}')
print()
print(f'Selected data: ')
df2[condition_for_selection2]
````

### Making a Deep Copy

````python
df2_copy = df2.copy()
df2_copy
````

````python
cities_new = ['Mumbai', 'Delhi', 'Chennai',
              'Kolkata', 'Bengalure', 'Hyderabad',
              'Pune', 'Ahmedabad', 'Guwahati']

df2_copy['new_city'] = cities_new
df2_copy
````

## Find and Replace

````python
criteria = df2['A'] < 0.2
print(criteria)
print()

# Changing the value  to 0 in col A and row selected bny criteria
df2.loc[criteria, 'A'] = 0
df2
````

`df2_copy` has a column `new_city` that `df2` doesn't have

````python
df2_copy.loc[df2_copy.city == 'Bengalure', ['city', 'new_city']] = 'Bengaluru'
df2_copy
````

````python
df2_copy.loc[:, 'city'] = 'Chennai'
df2_copy
````

## Dropping columns

````python
df2_copy.drop(['new_city'], axis=1)
````

## Statistical Measures

### Example 1

````python
random_state = np.random.RandomState(100)
random_series = pd.Series(random_state.rand(10))
````

Now that the series is created, we can try aggregation functions like `mean`, `std` etc.

````python
print(f'Mean: {random_series.mean()}')
print(f'Standard Deviation: {random_series.std()}')
print(f'Sum of all values: {random_series.sum()}')
````

## Groupby

Three stages
* Split - we split dataframe into multiple smaller dataframe based on the values of keys
* Apply - we apply desired aggregation/transformation on each dataframe.
* Combine - we combine results from apply state into a dataframe

split-apply-combine.png

````python
df4 = pd.DataFrame({'Key': ['A','B','C']*2,
                   'Data': range(6)})
df4
````

````python
df4.groupby("Key").sum()
````
